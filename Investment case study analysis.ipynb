## Checkpoint 3: Country Analysis

#### From the above,we conclude that venture type investment is better than others ,so we are going to perform operations based on venture type investments across countries

```python
#filtering venture type investments
master_funding_venture=master_df[master_df['funding_round_type']=='venture']
master_funding_venture
```

```python
Top_9_countries=master_funding_venture.groupby('country_code').sum().sort_values(by='raised_amount_usd',ascending=False)[:9]
Top_9_countries
```

#### From the wiki link of top countries speaking english,we found that the USA,GBR and IND match the order of top 3 countries present in our list that speak English,so we extract only those rows having these countries to perform further analysis

```python
master_funding_venture_by_country=master_funding_venture[(master_funding_venture['country_code']=='USA')|(master_funding_venture['country_code']=='GBR')|(master_funding_venture['country_code']=='IND')]
master_funding_venture_by_country
```

```python
sns.boxplot(x=master_funding_venture_by_country['country_code'],y=master_funding_venture_by_country['raised_amount_usd'])
plt.yscale('log')
plt.show()
```

```python
master_vf=master_funding_venture_by_country.groupby('country_code').describe()
master_vf
```

```python
# calculate summary statistics
GBR_funds_75,GBR_funds_25=master_vf.loc['GBR']['raised_amount_usd']['75%'],master_vf.loc['GBR']['raised_amount_usd']['25%']
IND_funds_75,IND_funds_25=master_vf.loc['IND']['raised_amount_usd']['75%'],master_vf.loc['IND']['raised_amount_usd']['25%']
USA_funds_75,USA_funds_25=master_vf.loc['USA']['raised_amount_usd']['75%'],master_vf.loc['USA']['raised_amount_usd']['25%']
# identify outliers
GBR_cut_off = (GBR_funds_75-GBR_funds_25) * 3
IND_cut_off = (IND_funds_75-IND_funds_25) * 3
USA_cut_off = (USA_funds_75-USA_funds_25) * 3

GBR_lower,GBR_upper = GBR_funds_25 - GBR_cut_off, GBR_funds_75 + GBR_cut_off
IND_lower,IND_upper = IND_funds_25 - IND_cut_off, IND_funds_75 + IND_cut_off
USA_lower,USA_upper = USA_funds_25 - USA_cut_off, USA_funds_75 + USA_cut_off

# identify outliers
data=master_funding_venture_by_country[master_funding_venture_by_country['country_code']=='GBR']['raised_amount_usd'] 
print('Total GBR: %d' % len(data))
GBR_outliers = [x for x in data if x < GBR_lower or x > GBR_upper]
GBR_outliers_removed = [x for x in data if x >= GBR_lower and x <= GBR_upper]
data=master_funding_venture_by_country[master_funding_venture_by_country['country_code']=='IND']['raised_amount_usd'] 
print('Total IND: %d' % len(data))
IND_outliers = [x for x in data if x < IND_lower or x > IND_upper]
IND_outliers_removed = [x for x in data if x >= IND_lower and x <= IND_upper]
data=master_funding_venture_by_country[master_funding_venture_by_country['country_code']=='USA']['raised_amount_usd'] 
print('Total USA: %d' % len(data))
USA_outliers = [x for x in data if x < USA_lower or x > USA_upper]
USA_outliers_removed = [x for x in data if x >= USA_lower and x <= USA_upper]
print('Identified GBR outliers: %d' % len(GBR_outliers))
print('Identified IND outliers: %d' % len(IND_outliers))
print('Identified USA outliers: %d' % len(USA_outliers))

# remove outliers

print('Non-outlier GBR observations: %d' % len(GBR_outliers_removed))
print('Non-outlier IND observations: %d' % len(IND_outliers_removed))
print('Non-outlier USA observations: %d' % len(USA_outliers_removed))
```

#### Dropping all the rows having the above outliers in each country 

```python
master_funding_venture_by_country.drop(master_funding_venture_by_country[(master_funding_venture_by_country['country_code']=='GBR')&(~master_funding_venture_by_country['raised_amount_usd'].isin(GBR_outliers_removed))].index, inplace = True)
master_funding_venture_by_country.drop(master_funding_venture_by_country[(master_funding_venture_by_country['country_code']=='IND')&(~master_funding_venture_by_country['raised_amount_usd'].isin(IND_outliers_removed))].index, inplace = True)
master_funding_venture_by_country.drop(master_funding_venture_by_country[(master_funding_venture_by_country['country_code']=='USA')&(~master_funding_venture_by_country['raised_amount_usd'].isin(USA_outliers_removed))].index, inplace = True)
```

```python
sns.boxplot(x=master_funding_venture_by_country['country_code'],y=master_funding_venture_by_country['raised_amount_usd'])
plt.yscale('log')
plt.show()
```
